```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

import Image as Image
import skimage.io as skio
import skimage as ski
import matplotlib.cm as cm
from scipy import ndimage
```

```python
from dropbox_path import dropbox_path
file_path = dropbox_path() + '\MicroCT\\2Dimages\S_20MMFOV_10UM_SLICE1_3X.TIF'
```

```python
im = skio.imread(file_path)
plt.imshow(im, cmap = cm.Greys_r)
plt.colorbar()
```

```python
from skimage.filter import canny
edges = canny(im/255.)
```

```python
segmented_image = ndimage.binary_fill_holes(edges)
plt.imshow(segmented_image, cmap = cm.Greys_r)
```
